### 横持ちと縦持ち

```python
import pandas as pd
import numpy as np

# 提供された品目リスト
item_list = [
    'まぐろ', 'あじ', 'いわし', 'かつお', 'さけ', 'さば', 'さんま', 'たい', 'ぶり', 'いか', 'たこ', 'えび', 'あさり', 'かき(貝)', 'ほたて貝',
    'キャベツ', 'ほうれんそう', 'はくさい', 'ねぎ', 'レタス', 'もやし', 'ブロッコリー', 'アスパラガス', 'さつまいも', 'じゃがいも', 'さといも', 'だいこん',
    'にんじん', 'ごぼう', 'たまねぎ', 'れんこん', 'ながいも', 'しょうが', 'えだまめ', 'さやいんげん', 'かぼちゃ', 'きゅうり', 'なす', 'トマト', 'ピーマン',
    '生しいたけ', 'えのきたけ', 'しめじ', 'カット野菜',
    'りんご', 'みかん', 'オレンジ', 'しらぬひ', '梨', 'ぶどう(デラウェア)', 'ぶどう(巨峰)', 'ぶどう(シャインマスカット)', '柿', '桃', 'すいか', 'メロン',
    'いちご', 'さくらんぼ', 'バナナ', 'キウイフルーツ', 'アボカド', 'パイナップル',
    '切り花(カーネーション)', '切り花(きく)', '切り花(バラ)'
]

# 横持ちデータを作成 (12ヶ月分のランダムな価格データ)
df_wide = pd.DataFrame(index=pd.date_range(start='2024-01-01', periods=12, freq='ME').strftime('%Y年%#m月'))
for item in item_list:
    df_wide[item] = np.random.randint(200, 1500, 12)

df_long = pd.melt(
    df_wide.reset_index(),
    id_vars=['index'],
    var_name='品目',
    value_name='価格'
)
df_long.rename(columns={'index': '月'}, inplace=True)

```

#### 横持ちデータ

横持ちデータは、列（カラム）に複数の項目が並んでいます。人間が見て直感的にわかりやすい形です。以下のdf_wideが横持ちデータです。
月ごとの各品目の価格が、それぞれの列に整理されています。

```python
df_wide
```

#### 縦持ちデータ

一方、縦持ちデータは、**列に「インデックス」「項目名」「値」**の3つを持ちます。
以下のdf_longが縦持ちデータです。
「月」「品目」「価格(円)」という3つの列に、すべての情報が整理されています。

```python
df_long
```

#### なぜ機械は縦持ちデータが好きか？

機械、特にプログラミングの処理では、縦持ちデータの方が扱いやすく、優れています。その理由は大きく3つあります。

#### 1. データの意味が明確になる
横持ちデータでは、各列が何のデータを表しているか、機械は自動では判断できません。人間がプログラムで**「この列はこういう意味だ」と指示**する必要があります。以下の例では、横持ちデータdf_wideから「まぐろ」の価格を取り出しています。このコードが動くのは、df_wideにまぐろという列が存在するという前提があるからです。

```python
# 横持ちデータから「まぐろ」の価格を抽出
price_maguro = df_wide['まぐろ']
print(f"まぐろの価格データ:\n{price_maguro}")
```

この方法だと、データに「ぶり」という新しい品目が追加された場合、機械はそれが価格データであることを認識できません。人間が**df_wide['ぶり']**というように、新しい列名を指定するようにプログラムを書き換える必要があります。

一方、縦持ちデータでは、「品目」という列にすべての品目名が、「価格」という列にすべての価格が格納されています。機械は、「品目という列の値が何であっても、対応する価格の列の値を取り出す」という一貫したルールで処理できます。

以下のコードでは、品目列の値が「まぐろ」である行の価格を取り出しています。このコードは、新しい品目（例えば「ぶり」）が追加されても、**df_long['品目'] == 'ぶり'**と条件を変えるだけで、プログラムの基本構造は一切変更する必要がありません。

```python
price_maguro_long = df_long[df_long['品目'] == 'まぐろ']
print(f"まぐろの価格データ:\n{price_maguro_long}")
```

#### 2. 汎用性が高い
横持ちデータで「魚介類だけの平均価格」や「野菜だけの平均価格」を計算したい場合、どの列が魚介類で、どの列が野菜なのかを手動で指定する必要があります。以下の例では、df_wideから「まぐろ」と「あじ」の列を選んで平均価格を計算しています。

```python
# 横持ちデータで、魚介類の特定の品目（まぐろ、あじ）の平均価格を計算
fish_items = ['まぐろ', 'あじ']
df_wide[fish_items].mean()
```

この方法では、もし新しい魚介類（「さんま」など）が加わった場合、fish_itemsのリストをプログラム内で手動で変更しなければなりません。データの種類が増えるたびに、プログラムを書き換える手間が発生します。

縦持ちデータでは、品目という列にすべての品目名が格納されているため、「魚介類」や「野菜」といった品目の分類情報を別の列として追加することができます。これにより、非常に柔軟なデータ処理が可能になります。
まずは、品目の分類情報を追加した新しい縦持ちデータdf_long_with_categoryを作成します。

```python
# 品目の分類情報を持つ縦持ちデータを作成
df_long_with_category = df_long.copy()

# 品目ごとの分類を辞書で定義
category_dict = {
    'まぐろ': '魚介', 'あじ': '魚介', 'いわし': '魚介', 'かつお': '魚介', 'さけ': '魚介',
    'キャベツ': '野菜', 'ほうれんそう': '野菜', 'はくさい': '野菜', 'ねぎ': '野菜',
    'りんご': '果物', 'みかん': '果物', 'オレンジ': '果物',
}

# '分類'列を追加
df_long_with_category['分類'] = df_long_with_category['品目'].map(category_dict)

# 新しいデータフレームの確認
df_long_with_category.head()
```

これと同じように横持ちデータに「分類」の情報を付与しようとすると、より複雑なデータ構造になってしまいます。たとえば、「まぐろ」に魚介という情報を付与する場合、「魚介_まぐろ」とするか、カラム名の２行にするかの作業が必要になります。この方法は、品目の種類が増えるほど列数も増え、非常に管理が難しくなります。

#### 3.集計や可視化のしやすさ

横持ちデータの場合<br>
横持ちデータでは、特定の条件でデータを集計したり可視化したりする際に、すべての列（カラム）を個別に指定する必要があります。

たとえば、「魚介類の平均価格」や「野菜の平均価格」を月ごとにグラフ化したい場合、どのカラムが魚介類で、どのカラムが野菜なのかをプログラムで手動で指定し、それぞれのグラフを別々に作成しなければなりません。この方法だと、品目が増えるたびに、グラフを作成するコードを書き換える手間が発生します。

```python
# 横持ちデータで特定の品目の価格をグラフ化
df_wide[['まぐろ', 'あじ', 'さけ']].plot()
plt.title('魚介類の価格推移')
plt.xlabel('年')
plt.ylabel('価格')
plt.show()
```

縦持ちデータでは、「分類」や「品目」といった列を使って、データを簡単にグループ化できます。これにより、集計や可視化の作業が非常にシンプルになります。たとえば、分類列を追加した縦持ちデータを使えば、魚介類、野菜、果物といったグループごとの平均価格を一度に計算し、グラフ化できます。

```python
# 縦持ちデータで分類ごとの平均価格をグラフ化
df_long_with_category.groupby(['月', '分類'])['価格'].mean().unstack().plot(
    title='分類ごとの平均価格推移'
)
```

このコードは、**「月と分類」でデータをグループ化し、それぞれの「価格(円)」**の平均を計算して、一度に複数の折れ線グラフを描画します。新しい分類のデータが追加されても、コードを変更する必要はありません。このように、縦持ちデータは集計の条件や可視化の対象を柔軟に変更できるため、データ分析の作業を効率化する上で非常に重要な形式です。

#### 4. 処理速度が速い
多くのプログラミング言語やデータ分析ツールは、縦持ちデータを効率よく処理するように設計されています。このため、縦持ちデータは横持ちデータよりも高速に動作します。

横持ちデータでは、データが増えるたびに新しい列を追加する必要があり、データの読み込みや計算に時間がかかることがあります。特に、データ数が非常に多い場合、その差は顕著になります。

縦持ちデータでは、行が増えても列の数は一定です。これは、機械がデータを読み込む際の構造が常に同じであることを意味し、高速な処理を可能にします。

具体的な処理例：大規模データで速度を比較
ここでは、より多くのデータを用意して、横持ちデータと縦持ちデータの処理速度の違いを実際に比べてみましょう。

まず、500,000行（約500,000品目×12ヶ月）の巨大な縦持ちデータと、それに対応する約500,000列の巨大な横持ちデータを作成します。

```python
import pandas as pd
import numpy as np
import time

# 巨大なデータセットを作成するための品目リスト
large_item_list = [f'品目_{i}' for i in range(500000)]

# 巨大な横持ちデータを作成
print("巨大な横持ちデータを作成中...")
start_time = time.time()
df_wide_large = pd.DataFrame(
    np.random.randint(200, 1500, (12, len(large_item_list))),
    index=pd.date_range(start='2024-01-01', periods=12, freq='ME').strftime('%Y年%#m月'),
    columns=large_item_list
)
wide_creation_time = time.time() - start_time
print(f"作成完了！ 所要時間: {wide_creation_time:.2f}秒")

# 巨大な縦持ちデータに変換
print("巨大な縦持ちデータに変換中...")
start_time = time.time()
df_long_large = df_wide_large.reset_index().rename(columns={'index': '月'})
df_long_large = pd.melt(df_long_large, id_vars=['月'], var_name='品目', value_name='価格(円)')
long_creation_time = time.time() - start_time
print(f"変換完了！ 所要時間: {long_creation_time:.2f}秒")
```

実行時間の比較<br>
次に、これらの巨大なデータセットを使って、**「各月の平均価格を計算する」**という同じ処理を実行し、それぞれの所要時間を計測します。

横持ちデータの処理速度<br>
横持ちデータでは、すべての列（品目）を選んで、その平均を計算します。これはdf_wide_large.mean(axis=1)という命令で実行します。

```python
import time

print("巨大な横持ちデータで価格の合計を計算中...")
start_time = time.time()
# まず行（月）ごとに合計し、次にその合計をすべて足し合わせる
# 例: 12行 × 100,000列のデータを合計
df_wide_large.sum().sum()
wide_process_time = time.time() - start_time
print(f"計算完了！ 所要時間: {wide_process_time:.4f}秒")
```

```python
print("巨大な縦持ちデータで価格の合計を計算中...")
start_time = time.time()
# '価格(円)'列の値をすべて合計する
# 例: 約120万行分のデータを合計
df_long_large['価格(円)'].sum()
long_process_time = time.time() - start_time
print(f"計算完了！ 所要時間: {long_process_time:.4f}秒")
```

#### 5. メモリとファイルサイズ

データ分析において、データが大きくなるとメモリ（コンピューターの作業用スペース）をどれだけ使うか、ファイルサイズがどれくらいになるか、という点も重要になります。横持ちデータは、縦持ちデータよりも多くのメモリとストレージ容量を消費する傾向があります。
なぜ横持ちデータはサイズが大きくなるのか？
これは、データ分析ツール（Pandasなど）がデータを管理する方法に理由があります。これらのツールは、データを列（カラム）ごとに同じデータ型で保存するように最適化されています。

横持ちデータでは、データが存在しない箇所（欠損値）があっても、そのための空のスペースがメモリ内に確保されます。たとえば、1000個の品目と12ヶ月分のデータがある場合、横持ちデータは12行 × 1000列のテーブルになります。もし、ある品目が特定の月にデータがなかったとしても、その場所は空のまま存在します。つまり、データがないにもかかわらず、そのためのメモリが消費されるのです。一方、縦持ちデータは「月」「品目」「価格(円)」という3つの列だけを使います。データが存在しない品目や月の組み合わせは、そもそも行として存在しないため、無駄なメモリを一切消費しません。このため、特に欠損値が多いデータを扱う場合、縦持ちデータの方がはるかに効率的です。
